```python
# Using heatmap ,show visualisation of given dataframe having null values
sb.heatmap(sr.isnull())
```

```python
# Fill none values with mean of total ratings paramentaly 
sr['Total ratings'].fillna(sr['Total ratings'].mean(),inplace=True)
```

```python
# Check dataset again that nan values is filled & no null vales pending
sr.isnull().sum()
```

```python
# Fetch all detail of Apps whose 5 star rating less than 70
sr[sr['5 star ratings']<70]
```

```python
# Fetch all detail of Apps whose 5 star rating less than 50 and unistalled app is also less than 50
sr[(sr['1 star ratings']<50)&(sr['Uninstalls']<=50)]
```

```python
# Fetch the count of all apps from datset 
sr['App'].value_counts()
```

```python
# Fetch the information of data whose app is temple Bun in dataset
sr[sr['App']=='Temple Bun']
```

```python
# using groupby() inbuilt function, Find the total sum of Total ratings of apps
sr.groupby('App')['Total ratings'].sum()
```

```python
# using groupby() inbuilt function, Find the total mean of Total ratings of apps
sr.groupby('App')['Total ratings'].mean()
```

```python
# using groupby() inbuilt function, Find the total minimum of Total ratings of apps
sr.groupby('App')['Total ratings'].min()
```

```python
# using groupby() inbuilt function, Find the total maximum of Total ratings of apps
sr.groupby('App')['Total ratings'].max()
```

```python
# Show relationship between Downloads and month using line graph & matplot library
plt.plot(sr['Downloads'],sr['Month'])
plt.scatter(sr['Downloads'],sr['Month'],color='blue',marker='*',s=100)# MArker - s=square ^=triangle, * = star
plt.show()
```

```python
# Draw Histogrma using inbuilt method hist() of pyplot
# Visualise histrogram (frequency graph)
plt.hist(sr['Downloads'],rwidth=0.8)
f=sr['Downloads'].value_counts()
plt.yticks(f)
plt.xlabel('Downloads')
plt.ylabel('Frequency')
plt.show()
```

```python
# How many Uninstalls in dataset present in graph
sr['Uninstalls'].value_counts().plot(kind='bar',color=['pink','skyblue'],edgecolor='black')
f=sr['Uninstalls'].value_counts() 
plt.yticks(f)
plt.xlabel('Uninstalls')
plt.ylabel('Frequency')
plt.show()
```

```python
# Using pie graph show all apps percentage 
sr['App'].value_counts().plot(kind='pie',colors=['lightgrey','skyblue','lightpink','cornflowerblue'],autopct='%d%%')
plt.show()
```

```python
# Export apps which got one star rating and compare using graph
sb.countplot(data=sr,x='1 star ratings',hue='App')
f=sr['1 star ratings'].value_counts()
plt.yticks(f)
plt.show()
```

```python
# Using Boxplot , Find the outlier in Total rating
sb.boxplot(data=sr,x='Total ratings')
plt.show()
```

```python
# Sort from dataset whose total rating is more than 900
sr[sr['Total ratings']>900]
```

```python
# Create a displot using seaborn library of total ratings from dataset 
sb.distplot(sr['Total ratings'])
plt.show()
```

```python
# Draw a scatterplot from given datset of Downloads and Uninstalls
sb.scatterplot(data=sr,x='Downloads',y='Uninstalls')
plt.show()
```

```python
# # Find the correlation between variables from given dataset
sr.corr()
```

```python
# Using heatmap with label for the given dataset , find the correlation between variables
sb.heatmap(sr.corr(),annot=True) 
plt.show()
```

```python
# Using Scatterplot for given dataset, provide prediction for 5 star ratings,Downloads, Uninstalls
x=['5 star ratings','Downloads', 'Uninstalls']
plt.figure(figsize=(8,7))
for i,col in enumerate(x):
    print(col)
    plt.subplot(2,2,i+1)# row , column
    sb.scatterplot(data=sr,x=col,y='Total ratings')
plt.show()
```
